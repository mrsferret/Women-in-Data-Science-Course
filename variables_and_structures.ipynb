## What is a variable? <a class="anchor" id="chapter_what_is_a_variable"></a>

Variables are containers for storing data values once you've loaded them into your program. They can then be saved, referenced or manipulated. They also provide a way of labelling data with a descriptive name, so our programs can be understood by ourselves and the computer. 

It is helpful to think of variables as containers that hold information in them. More specifically, they are reserved memory locations to store values. This means that when you create a variable in Python, you reserve some space in memory for you data. 

A good analogy for this could be to think of packing moving boxes: you would store a certain type of object in each individual box, and label that box appropriately so that everyone knows what is in it. In this analogy, the box itself would be a variable.

### Strings <a class="anchor" id="chapter_strings"></a>

In Python 3, a string represents a sequence of characters. 

Strings are amongst the most popular types in Python. We can create them simply by enclosing characters in quotes. Python treats single quotes the same as double quotes. 

We can use this opportunity to create our **first variable**. There are a few different steps to this, so let's type it out and go through it. 


```python
# make our first string!
my_string = "Hello World!"

# print out our first string
my_string
```

<br>

So, what are we doing here? 

1. The first part of this is the `my_string` part. This is us defining our variable. This name is simply a holder - we have called it `my_string` because it contains our first string. 

2. The second part is the `=` sign. This is how we assign information to our variable. So we are saying "set `my_string` to `=`..."

3. The third part is where we enter our string. In this case, we are putting in our characters `Hello World`. This is a common string used in programming languages to set up strings. 

4. The forth part we have done here is to print the data out. We can do this, by typing our variable name. The default in programming is not to print out variables, because they could contain a lot of information and printing them out might be computationally heavy. By printing our variable out, it allows us to see what is inside it. 

And voila - we've made our first **string variable** in Python!

<br>

### Boolean <a class="anchor" id="chapter_boolean"></a>

A boolean expression (or logical expression) evaluates to one of two states: `True` or `False`. Python provides the boolean type that can be either set to False or True.

The capitalisation is important: only `True` and `False` are of type `bool`.

Expressions in Python are often evaluated in Boolean context, meaning they are interpreted to represent truth or falsehood. Booleans are a concept that exists in every programming language. A boolean represents the idea of "true" or "false". When you are writing a program, there are often circumstances where you want to execute different code in different situations. Booleans enable our code to do just that.




A boolean value is returned as a result of some kind of comparison between variables. For example, the following code would store a boolean value of `False` in the `have_same_name variable` after using the equality comparison operator, the `==` symbol.

```python
# create two name vectors
my_name = "Steph"
instructor_name = "Aileen"

# check if they are the same
my_name == instructor_name
```

Booleans are an important tool in any programming language. Using boolean logic, you can do more complex things like write loops and functions. This lets your code can react to data inside your program, and carry out different instructions under different circumstances. This is a bit more advanced than what we are doing.

## Declaring variables <a class="anchor" id="chapter_declaring_variables"></a>

Python is a *dynamically typed language*. One of the implications of this is that we don't have to specify what type a variable is when we declare it; we simply use an equals sign to assign a value to a variable name.

In addition, we don't need to declare variables before assigning them. We can just use them straight away.

As we saw before, we can do something like this:

```python
my_name = "Steph"
my_name
```

Because Python is dynamically typed, there's nothing to stop the type of a variable from changing:

```python
my_name = "Steph"
my_name = 5
my_name
```

However, because Python is also _strongly_ typed, it will throw an error if we try to combine two incompatible types. We can't add a `str` and an `int` together, for example:

```python
# my_variable = "5" + 10
```


<br>




<br>


Python errors can be pretty confusing to read and make sense of, but the key bit you're looking at is the line noting `TypeError: can only concatenate str (not "int") to str`. 

This tells you what the error is, and can often be a good guide about what is wrong. Here, it is telling us it can't concatenate an integer to a string, which is exactly what is wrong!

## Lists <a class="anchor" id="chapter_lists"></a>

In Python, lists are ordered collections of items that allow for easy use of a set of data. Lists are great because they can contain multiple items. We can declare a list in much the same way as we would declare another variable. We use a variable name and the equals sign. The main difference is that we then place the values we want to hold in the list inside square brackets`[]`. 


```python
# define a string list
my_list = ["spam", "ham", "eggs", "ham"]

# define a number list 
my_number_list = [1, 3, 4, 6, 10]
```

In Python, lists are a versatile data type that can contain multiple different data types within the same square brackets. The possible data types within a list include numbers, strings, other objects, and even other lists. The values in a list also do not need to be unique, which makes them perfect for storing data. 

*It is good practice to put a space between the comma and the next value when making a list.*



### Accessing a list <a class="anchor" id="chapter_accessing_lists"></a>

Python list elements are ordered by index, a number referring to their placement in the list. We can access the various elements of the list by _index_ - the position of the element in the list. If we take our first list, we can start to work with it. 

```python
my_list
```

One thing that must be made clear before we start indexing, is that Python is a **zero indexed language**. This means that it starts counting positions at zero. This can be somewhat confusing when you're not used to programming - you'd be tempted to count the first thing in the list as holding position '1'. For example, in `my_list`, if I asked you which position `spam` was in, you might say one. But in fact, what you actually need to say is `0`, because Python starts counting from zero. 

You can test this out by **indexing**. You do this by putting the name of your list in, and then using square brackets `[]` to input the position. Therefore, if we put 0 in as the position, it will return what is in position 0. 

```python
my_list[0]
```

There you go, we have got an output `spam`, which tells us the value `spam` is in position 0 in our list. 

We can also try another position:

```python
my_list[2]
```

We can also access counting from the back of the list. Confusingly, the end position is denoted by `-1`, because this is using a negative offset value from the end of the list. 


```python
my_list[-1]
```

You can also do what is called slicing, which is where you use the `:` operator to take multiple bits of a list.

```python
my_list[1:3]
```

The first number is the position at start of our slice i.e. "start at position 2" (which is index 1). The second number is the position **just before** we want to finish our slice i.e. "end just before position 4" (which is index 
3). 

When slicing a list, a new list is returned, so if the slice is saved and then altered, the original list remains the same.

<br>

## Dictionaries <a class="anchor" id="chapter_dictionaries"></a>

It's possible in Python to create a list containing items of different types. For example, we might want to collect data about a person in a list, such as their name, age, job title and whether or not they are an instructor. 

If we want to associate values with easier means of identifying them, we can use Python's `Dictionary`.**A Python dictionary is an ordered collection of items. It contains data as a set of 'key:value' pairs.**

We can set up a dictionary below like follows:

```python
person = {"name": "Steph", "age": 32, "job" : "Instructor"}
```

While we used square brackets to denote a list, we can use curly brackets to declare a dictionary. Values in a dictionary are referred to by a _key_. This is what you see appearing before the colon symbol `:` in the above code. 

Usually, the keys for our dictionary will be strings, but they can also be numeric. Python allows the values in a dictionary to be any type  string, integer, a list, another dictionary, boolean, etc.

<br>

### Indexing a dictionary <a class="anchor" id="chapter_indexing_dictionary"></a>

To get information back out of a dictionary, we can use square-bracket notation, similar to a list. Values in a Python dictionary can be accessed by placing the key within square brackets next to the dictionary. Values can be written by placing key within square brackets next to the dictionary and using the assignment operator (=). 


```python
person["name"]
```

If we want to assign a new key and value to `user`, we can do so like this:

```python
person["email"] = "steph@codeclan.com"

person
```

## Differences: Lists and Dictionaries <a class="anchor" id="chapter_lists_and_dictionaries"></a>

<br>
It's hard to remember the difference between lists and dictionaries. The main things are:

    * Lists use [ ] brackets
    * Dictionaries use { } brackets
    
    * Lists hold ordered collections of anything
    * Dictionaries hold key:value pairs
    
    
This is the key part to remember. You can actually combine the two, so you have a list within a dictionary. This lets you hold multiple things:

```python
person = {
    "name": "Steph",
    "age": 32,
    "pets": ["Oscar", "Maia", "Archie"]
}

person
```

If you wanted to get something out of this, you index just in the same way you always do: with the `[ ]` brackets.

```python
person["pets"][2]
```
